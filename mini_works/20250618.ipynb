```python
from sklearn.datasets import load_breast_cancer
```

```python
raw_data = load_breast_cancer()
```

```python
import pandas as pd
import numpy as np
```

```python
raw_data.keys()
```

```python
df_x = pd.DataFrame(data=raw_data['data'] ,  columns=raw_data['feature_names'])
```

```python
target = raw_data['target']
```

```python
raw_data['target_names']
```

```python
df_x
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, target, test_size=0.2, random_state=42)
```

```python

st = StandardScaler()
st.fit(x_train)  # 평균과 표준편차 계산

x_train_s =st.transform(x_train)
x_test_s = st.transform(x_test)
```

```python

st = StandardScaler()
st.fit(x_train)  # 평균과 표준편차 계산
st.fit(x_test)  # 평균과 표준편차 계산

x_train_s2 =st.transform(x_train)
x_test_s2 = st.transform(x_test)
```

```python
minmax = MinMaxScaler()
minmax.fit(x_train)

x_train_m = minmax.transform(x_train)
x_test_m = minmax.transform(x_test)
```

```python
pd.DataFrame(x_train_s)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrg = LogisticRegression()
lrg.fit(x_train_s, y_train)
```

```python
y_test == lrg.predict(x_test_s)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(lrg.predict(x_test_s), y_test
               )
```

```python
lrg_m = LogisticRegression()

lrg_m.fit(x_train_m, y_train)
```
